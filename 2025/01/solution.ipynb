## Part 1

```python
lines = """
L68
L30
R48
L5
R60
L55
L1
L99
R14
L82""".strip().split()

with open("input.txt", "r") as f:
    lines = f.readlines()
```

```python
items = [ int(el[1:]) if el[0] == "R" else -1 * int(el[1:]) for el in lines ]

position = 50
counter = 0

for rotations in items:
    position = position + rotations % 100 # <- removes the negative side

    if position > 99:
        position = position - 100

    if position == 0:
        counter += 1

print(counter)
```

## Part 2

```python
lines = """
L68
L30
R48
L5
R60
L55
L1
L99
R14
L82""".strip().split()

with open("input.txt", "r") as f:
    lines = f.readlines()

rotations = [ int(el[1:]) if el[0] == "R" else -1 * int(el[1:]) for el in lines ]

position = 50
counter = 0

for clicks in rotations:
    direction = 1
    if clicks < 0:
        direction = -1
    for _ in range(abs(clicks)):
        position += direction
        if position > 99:
            position = 0
        if position < 0:
            position = 99
        if position == 0:
            counter += 1

print(counter)
```

## Part 1

```python
data = ("11-22,95-115,998-1012,"
        "1188511880-1188511890,"
        "222220-222224,1698522-1698528,"
        "446443-446449,38593856-38593862,"
        "565653-565659,824824821-824824827,"
        "2121212118-2121212124")

with open("input.txt", "r") as f:
    data = f.read()

items = data.split(",")
```

```python
def get_range(id_range:str):
    start, end = id_range.split("-")
    return range(int(start), int(end)+1)

def is_invalid(number:int) -> bool:
    string = str(number)
    if string[0:len(string)//2] == string[len(string)//2:]:
        return True
    return False

sum_of_invalids = 0

for item in items:
    for number in get_range(item):
        if is_invalid(number):
            sum_of_invalids += number

print(sum_of_invalids)
```

## Part 2

```python
data = ("11-22,95-115,998-1012,"
        "1188511880-1188511890,"
        "222220-222224,1698522-1698528,"
        "446443-446449,38593856-38593862,"
        "565653-565659,824824821-824824827,"
        "2121212118-2121212124")

with open("input.txt", "r") as f:
    data = f.read()

items = data.split(",")
```

```python
import functools

def get_parts(string:str, n: int):
    return tuple([string[i:i+n] for i in range(0, len(string), n)])

@functools.cache
def is_invalid_v2(number:int) -> bool:
    string = str(number)
    for n in range(1, len(string) // 2 + 1):
        parts = get_parts(string, n)
        if all(x == parts[0] for x in parts):
            return True
    return False

sum_of_invalids = 0

for item in items:
    for number in get_range(item):
        if is_invalid_v2(number):
            sum_of_invalids += number

print(sum_of_invalids)
```

## Part 1

```python
lines = """
987654321111111
811111111111119
234234234234278
818181911112111
""".strip().split()

with open("input.txt", "r") as f:
    lines = f.readlines()
```

```python
def get_batteries(stack: str, number_of_baterries:int=2) -> int:
    batteries = list(stack.strip())
    batteries_index = -1

    places = [-1] * number_of_baterries

    for place_idx in range(0, len(places)):
        for bt_idx in range(batteries_index+1, len(batteries)-(len(places)-place_idx-1)):

            if int(batteries[bt_idx]) > places[place_idx]:
                places[place_idx] = int(batteries[bt_idx])
                batteries_index = bt_idx

    return int("".join([str(n) for n in places]))

sum_of_jolts = 0
for line in lines:
    jolts = get_batteries(line)
    sum_of_jolts += jolts

print(sum_of_jolts)
```

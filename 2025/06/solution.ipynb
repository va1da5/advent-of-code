## Part 1

```python
import math

data = """
123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
""".strip()

with open("input.txt", "r") as f:
    data = f.read()
```

```python
def get_items(row:str):
    columns = row.strip().split()
    try:
        return [int(n) for n in columns]
    except ValueError:
        return columns

def parse(data:str):
    rows = []
    for line in data.strip().split("\n"):
        rows.append(get_items(line))

    return list(zip(*rows))


total_sum = 0

for column in parse(data):
    column = list(column)
    operator = column.pop()
    if operator == "+":
        total_sum += sum(column)
    else:
        total_sum += math.prod(column)

total_sum
```

## Part 2

```python
data = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """

with open("input.txt", "r") as f:
    data = f.read()
```

```python
def get_matrix(data:str):
    matrix = []
    for line in data.split("\n"):
        matrix.append(list(line))
    return list(zip(*matrix))

total_sum = 0

def parse_cephalopod_math(matrix):

    matrix.append([" "] * len(matrix[0]))
    
    columns = []
    parameter = ""
    column = []
    for row in matrix:
        # print(row)
        if "*" in row or "+" in row:
            parameter = row[-1]
            number = int("".join(row[:-1]))
            column.append(number)
            continue

        if all([el == " " for el in row]):
            
            columns.append(
                column + [parameter]
            )
            column = []
            continue

        column.append(int("".join(row)))

    return columns


total_sum = 0

for column in parse_cephalopod_math(get_matrix(data)):
    column = list(column)
    operator = column.pop()
    if operator == "+":
        total_sum += sum(column)
    else:
        total_sum += math.prod(column)

total_sum
```

## Part 1

```python
data = """
.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
""".strip()

with open("input.txt", "r") as f:
    data = f.read()
```

```python
def get_matrix(data:str):
    rows = []
    for line in data.strip().split("\n"):
        rows.append(list(line))

    return rows

def count_splits(matrix):
    split_counter = 0
    
    for y in range(1,len(matrix)):
        for x in range(len(matrix[0])):
            if matrix[y-1][x] == "S":
                matrix[y][x] = "|"

            if matrix[y][x] == "^" and matrix[y-1][x] == "|":
                matrix[y][x+1] = "|"
                split_counter += 1
                if matrix[y][x-1] != "|":
                    matrix[y][x-1] = "|"

            if matrix[y][x] == "." and matrix[y-1][x] == "|":
                matrix[y][x] = "|"                

    return split_counter

count_splits(get_matrix(data))
```

## Part 2

```python
def render_matrix(matrix) -> str:
    print("\n".join(["".join(line) for line in matrix]))

def calc_timelines(matrix):
    timelines = [0] * len(matrix[0])
    
    for y in range(1,len(matrix)):
        for x in range(len(matrix[0])):
            
            if matrix[y][x] == "." and matrix[y-1][x] == "S":
                matrix[y][x] = "|"
                timelines[x] += 1


            if matrix[y][x] == "^" and matrix[y-1][x] == "|":
                matrix[y][x+1] = "|"
                matrix[y][x-1] = "|"
                timelines[x-1] += timelines[x]
                timelines[x+1] += timelines[x]
                timelines[x] = 0

            if matrix[y][x] == "." and matrix[y-1][x] == "|":
                matrix[y][x] = "|"

    return sum(timelines)

calc_timelines(get_matrix(data))
```

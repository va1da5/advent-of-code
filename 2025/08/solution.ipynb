```python
import math
import functools
```

```python
data = """
162,817,812
57,618,57
906,360,560
592,479,940
352,342,300
466,668,158
542,29,236
431,825,988
739,650,466
52,470,668
216,146,977
819,987,18
117,168,530
805,96,715
346,949,466
970,615,88
941,993,340
862,61,35
984,92,344
425,690,689
"""

# with open("input.txt", "r") as f:
#     data = f.read()
```

```python
def get_positions(data:str):
    positions = []
    for line in data.strip().split("\n"):
        positions.append(tuple([int(n) for n in line.split(",")]))
    return positions

@functools.cache
def euclidean_distance(a, b):
    ax,ay,az = a
    bx,by,bz = b
    return math.sqrt(pow(ax-bx, 2) + pow(ay-by, 2) + pow(az-bz, 2))
    # return math.sqrt(pow(cx - fx, 2) + pow(cy - fy, 2))

    
positions = get_positions(data)
```

```python
def get_distances(positions):
    distances = []
    for a_idx in range(len(positions)):
        for b_idx in range(a_idx+1, len(positions)):
            a = positions[a_idx]
            b = positions[b_idx]
            distances.append((a, b, euclidean_distance(a,b)))

    # return distances
    return sorted(distances, key=lambda x: x[2])

get_distances(get_positions(data))
```

```python
def count_connections(circuits):
    count = 0
    for  connections in circuits:
        count += len(connections) - 1

    return count

def get_circuits(distances, max_connections=10):
    circuits = []

    def make_new_circuit(a, b):
        new_circuit = set()
        new_circuit.add(a)
        new_circuit.add(b)
        circuits.append(new_circuit)
        
        
    for a, b, _ in distances:
        connections = count_connections(circuits)
        print(f" {connections=}")
        
        if  connections>= max_connections:
            return circuits

        if not len(circuits):
            make_new_circuit(a,b)
            print(f" new circuit({(a,b)})")
        
        for circuit_idx in range(len(circuits)):
            if a in circuits[circuit_idx] and b in circuits[circuit_idx]:
                print(f" {a} & {b} exists in [{circuits[circuit_idx]}]")
                break
            if a in circuits[circuit_idx] or b in circuits[circuit_idx]:
                print(f" [{circuits[circuit_idx]}] + ({a}) + ({b})")
                circuits[circuit_idx].add(a)
                circuits[circuit_idx].add(b)
                break
        else:
            make_new_circuit(a,b)
            print(f" append({(a,b)})")

    return []

def calculate_result(circuits):
    return math.prod([len(circuit) for circuit in circuits[:3]])
    

distances = get_distances(get_positions(data))

calculate_result(get_circuits(distances, 1000))
```
